#### this helps to convert ciphertext of password into 8-bit binary values which further converted into ascii value and it used in des.cpp file to generate ciphertext

```python
s = "rsqonrdgspfqikrghpsrnlshoolpgmmj"
    
    
def str_to_bitarray(text):
    array = list()
    # Remember Our text is from f(0000) to u(1111)
    for i in text:
        temp = ord(i) - ord('d')
        temp = "".join(bin(temp))[2:].zfill(4)
        array.extend([int(x) for x in list(temp)])
    return array

a=str_to_bitarray(s)
c=8
for i in a:
    if(c==0):
        print("\n")
        c=8
    print(i,end="")
    c-=1
```
